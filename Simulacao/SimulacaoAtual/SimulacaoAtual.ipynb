# Simulação do modelo atual

Esse documento tem como objetivo realizar e documentar uma simulação do sistema atual, evidenciando seu comportamento e desempenho, bem como analisar possíveis pontos de melhoria, riscos e oportunidades.

## Importação das bibliotecas necessárias

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Lógica de simulação

```python
def simulacao(estado1,estado2,probabilidade=0.65):
  if np.random.rand() <= probabilidade:
    return estado1
  else:
    return estado2

estado1 = "Categorização correta"
estado2 = "Categorização incorreta"

simulacoes = []
```

## Simulação da categorização de 24 compras distintas

```python
compra_atual = 1

for a in range(10000):
  resultado = simulacao(estado1,estado2,0.65)
  simulacoes.append(resultado)
  print(f"Compra {compra_atual} - {resultado}")
  compra_atual += 1
```

### Exibição dos resultados da simulação

```python
count_acertos = simulacoes.count(estado1)
count_erros = simulacoes.count(estado2)

porcentagem_acertos = (count_acertos / 10000) * 100
porcentagem_erros = (count_erros / 10000) * 100

print(f"Porcentagem de categorizações corretas: {porcentagem_acertos}%")
print(f"Porcentagem de categorizações incorretas: {porcentagem_erros}%")
```

```python
estados = [estado1,estado2]
contagens = [count_acertos,count_erros]
plt.bar(estados,contagens)
plt.ylim(0,10000)
plt.show
```
