```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
import pandas as pd


def transformar_chaves(objeto_json):
    # Mapeia as chaves do objeto JSON
    mapeamento_chaves = {
        "supplier_name": "Supplier Name",
        "region": "Region",
        "country_name": "Country Name",
        "strategic_region": "Strategic Region",
        "business_unit": "Business Unit",
        "legal_entity": "Legal Entity",
        "cost_center_base": "Cost Center (Base Level)",
        "invoice_source": "Invoice Source",
        "gl_description": "GL Description",
        "amount_usd": "Amount USD"
    }
    # Cria um novo dicionário com as chaves transformadas
    objeto_transformado = {mapeamento_chaves[chave]: valor for chave, valor in objeto_json.items()}

    return objeto_transformado

auau = {'Supplier Name': 'ACUDIR SA', 'Region': 'LATAM', 'Country Name': 'Argentina', 'Strategic Region': 'LATAM', 'Business Unit': 'Technology/Telecom', 'Legal Entity': 'Facebook Argentina S.R.L.', 'Cost Center (Base Level)':'Global Security Group', 'Invoice Source': 'OCR', 'GL Description': 'Temporary services', 'Amount USD': 1200}


def padronizacao(dado):
    df = pd.read_excel("/content/drive/MyDrive/database/LATAM-Data.xlsx")
    df = df.drop(['Normalized Supplier Name', 'Parent Supplier Name', 'Product', 'Requestor Name','Unnamed: 0', 'Preparer Name', 'PO Number', 'Project', 'Level 3', 'Invoice ID', 'Invoice Number', 'Cost Center (Level 6)', 'GL Desc (Level 6)', 'Cost Center', 'Month, Day, Year of Payment Date', 'Cost Center (Level 4)', 'Cost Center (Level 5)', 'GL Desc (Level 4)', 'GL Desc (Level 5)'], axis=1)
    df = df.dropna(subset=['Business Unit'])
    df = df.loc[df['Level 1'] != 'Uncategorized']
    colunas_numericas = df.select_dtypes(include=['float64', 'int64']).columns
    scaler = StandardScaler()
    df[colunas_numericas] = scaler.fit_transform(df[colunas_numericas])
    df = df.drop('Level 2', axis=1)
    data_sem_level1 = df.drop('Level 1', axis=1)
    df_encoded = pd.get_dummies(data_sem_level1)
    colunas_treinamento = df_encoded.columns
    linha = pd.DataFrame([dado])
    linha = linha.dropna(subset=['Business Unit'])
    colunas_numericas = linha.select_dtypes(include=['float64', 'int64']).columns
    linha[colunas_numericas] = scaler.fit_transform(linha[colunas_numericas])
    # Aplica One-Hot Encoding nas colunas categóricas
    linha_encoded = pd.get_dummies(linha)
    # Identifica as colunas faltantes
    colunas_faltantes = set(colunas_treinamento) - set(linha_encoded.columns)
    # Converte colunas_faltantes de um set para uma lista
    colunas_faltantes_lista = list(colunas_faltantes)
    # Cria um novo DataFrame para as colunas faltantes
    df_colunas_faltantes = pd.DataFrame(0, index=np.arange(len(linha)), columns=colunas_faltantes_lista)
    # Concatena o DataFrame das colunas faltantes
    linha_encoded = pd.concat([linha_encoded, df_colunas_faltantes], axis=1)
    # Garante a mesma ordem de colunas como no treinamento
    linha_encoded = linha_encoded.reindex(columns=colunas_treinamento)

    with open('modelo', 'rb') as arquivo:
      modelo = pickle.load(arquivo)
      resposta_level1 = modelo.predict(linha_encoded)
      print(resposta_level1)

    linha_encoded['Level 1'] = resposta_level1
    df = pd.read_excel("/content/drive/MyDrive/database/LATAM-Data.xlsx")
    df = df.drop(['Normalized Supplier Name', 'Parent Supplier Name', 'Product', 'Requestor Name','Unnamed: 0', 'Preparer Name', 'PO Number', 'Project', 'Level 3', 'Invoice ID', 'Invoice Number', 'Cost Center (Level 6)', 'GL Desc (Level 6)', 'Cost Center', 'Month, Day, Year of Payment Date', 'Cost Center (Level 4)', 'Cost Center (Level 5)', 'GL Desc (Level 4)', 'GL Desc (Level 5)'], axis=1)
    df = df.dropna(subset=['Business Unit'])
    df = df.loc[df['Level 1'] != 'Uncategorized']
    colunas_numericas = df.select_dtypes(include=['float64', 'int64']).columns
    df[colunas_numericas] = scaler.fit_transform(df[colunas_numericas])
    df = df.drop('Level 2', axis=1)
    df_encoded = pd.get_dummies(df)
    colunas_treinamento = df_encoded.columns
    linha = pd.DataFrame([dado])
    linha = linha.dropna(subset=['Business Unit'])
    colunas_numericas = linha.select_dtypes(include=['float64', 'int64']).columns
    linha[colunas_numericas] = scaler.fit_transform(linha[colunas_numericas])
    linha_encoded = pd.get_dummies(linha)
    colunas_faltantes = set(colunas_treinamento) - set(linha_encoded.columns)
    # Converte colunas_faltantes de um set para uma lista
    colunas_faltantes_lista = list(colunas_faltantes)
    # Cria um novo DataFrame para as colunas faltantes
    df_colunas_faltantes = pd.DataFrame(0, index=np.arange(len(linha)), columns=colunas_faltantes_lista)
    # Concatena o DataFrame das colunas faltantes
    linha_encoded = pd.concat([linha_encoded, df_colunas_faltantes], axis=1)
    # Garante a mesma ordem de colunas como no treinamento
    linha_encoded = linha_encoded.reindex(columns=colunas_treinamento)

    with open('modelo_2', 'rb') as arquivo:
        modelo_level2 = pickle.load(arquivo)
    resposta_level2 = modelo_level2.predict(linha_encoded)

    linha_encoded['Level 2'] = resposta_level2

    df = pd.read_excel("/content/drive/MyDrive/database/LATAM-Data.xlsx")
    df = df.drop(['Normalized Supplier Name', 'Parent Supplier Name', 'Product', 'Requestor Name','Unnamed: 0', 'Preparer Name', 'PO Number', 'Project', 'Invoice ID', 'Invoice Number', 'Cost Center (Level 6)', 'GL Desc (Level 6)', 'Cost Center', 'Month, Day, Year of Payment Date', 'Cost Center (Level 4)', 'Cost Center (Level 5)', 'GL Desc (Level 4)', 'GL Desc (Level 5)'], axis=1)
    df = df.dropna(subset=['Business Unit'])
    df = df.loc[df['Level 1'] != 'Uncategorized']
    colunas_numericas = df.select_dtypes(include=['float64', 'int64']).columns
    df[colunas_numericas] = scaler.fit_transform(df[colunas_numericas])
    df = df.drop('Level 3', axis=1)
    df_encoded = pd.get_dummies(df)
    colunas_treinamento = df_encoded.columns
    linha = pd.DataFrame([dado])
    linha = linha.dropna(subset=['Business Unit'])
    colunas_numericas = linha.select_dtypes(include=['float64', 'int64']).columns
    linha[colunas_numericas] = scaler.fit_transform(linha[colunas_numericas])
    linha_encoded = pd.get_dummies(linha)
    colunas_faltantes = set(colunas_treinamento) - set(linha_encoded.columns)
    # Converte colunas_faltantes de um set para uma lista
    colunas_faltantes_lista = list(colunas_faltantes)
    # Cria um novo DataFrame para as colunas faltantes
    df_colunas_faltantes = pd.DataFrame(0, index=np.arange(len(linha)), columns=colunas_faltantes_lista)
    # Concatena o DataFrame das colunas faltantes
    linha_encoded = pd.concat([linha_encoded, df_colunas_faltantes], axis=1)
    # Garante a mesma ordem de colunas como no treinamento
    linha_encoded = linha_encoded.reindex(columns=colunas_treinamento)
    with open('modelo_3', 'rb') as arquivo:
        modelo_level3 = pickle.load(arquivo)
    resposta_level3 = modelo_level3.predict(linha_encoded)

    print('Level 1:', resposta_level1, 'Level 2:', resposta_level2, 'Level 3:', resposta_level3)
    return resposta_level1, resposta_level2, resposta_level3


resposta = padronizacao(auau)

```

```python
import paho.mqtt.client as mqtt
import json

# Definição das informações do broker HiveMQ
broker_host = "broker.hivemq.com"
broker_port = 1883  # Porta para MQTT
topic_subscribe = "/purchase"  # Tópico para inscrição
topic_publish = "/purchase-rendered"  # Tópico para publicação

def on_connect(client, userdata, flags, rc):
    print(f"Conectado com código de resultado {rc}")
    # Inscrição no tópico desejado após a conexão
    client.subscribe(topic_subscribe, qos=0)

def on_message(client, userdata, msg):
    try:
        # Decodificar a carga útil (payload) como um objeto JSON
        json_data = json.loads(msg.payload.decode())
        json_data["amount_usd"] = float(json_data["amount_usd"])

        # Imprimir ou processar os dados recebidos conforme necessário
        print(f"Mensagem recebida {msg.topic}: {json_data}")

        # Publicar a mensagem no novo tópico
        client.publish(topic_publish, payload=json.dumps(json_data), qos=0)

        json_transformado = transformar_chaves(json_data)

        resultado = padronizacao(json_transformado)
        print(resultado)

    except json.JSONDecodeError as e:
        print(f"Erro ao decodificar JSON: {e}")

def on_disconnect(client, userdata, rc):
    print(f"Desconectado com código de resultado {rc}")

# Criação de uma instância do cliente MQTT
client = mqtt.Client()

# Associação das funções de callback ao cliente
client.on_connect = on_connect
client.on_message = on_message
client.on_disconnect = on_disconnect

# Conexão ao broker HiveMQ
client.connect(broker_host, broker_port, keepalive=60)

# Mantendo a conexão ativa
client.loop_start()

# Mantendo o script em execução para permitir a recepção de mensagens
try:
    while True:
        pass
except KeyboardInterrupt:
    # Lidando com a interrupção do teclado (Ctrl+C)
    print("Desconectando...")
    client.disconnect()
    client.loop_stop()
    print("Desconectado com sucesso.")
```
